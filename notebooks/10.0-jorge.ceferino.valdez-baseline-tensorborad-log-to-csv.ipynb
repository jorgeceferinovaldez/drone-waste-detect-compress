```python
import os
import sys
import csv
from pathlib import Path
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
```

```python
# Me aseguro de que el directorio raíz del proyecto esté en el sys.path
project_root = Path(os.path.abspath("")).parent

# Añado el directorio raíz al sys.path si no está ya presente
if project_root not in sys.path:
    sys.path.append(str(project_root))
```

```python
# Importo las funciones de configuración
from src.config import reports_dir, load_config
```

```python
# Cargo la configuración 
config = load_config()
```

```python
# Defino directorio base de logs para tensorboard
log_base_dir = reports_dir()/'logs'
```

```python
nombre_modelo = 'convolutional_autoencoder_model_final_best_hyperparameters_2_2'
```

```python
# Ruta del directorio de logs
#logdir = "../reports/logs/compressai_cheng2020_anchor_2_2_BIS"
log_dir = log_base_dir / nombre_modelo
log_dir = str(log_dir)
```

```python
print(f"Directorio de logs: {log_dir}")
```

```python
# Inicializar el EventAccumulator para leer los logs
event_acc = EventAccumulator(log_dir)
event_acc.Reload()
```

```python
# Extraer los tags de scalars disponibles en los logs
tags = event_acc.Tags()['scalars']
```

```python
output_dir = reports_dir()/'tensorboard_metrics_csv'/nombre_modelo
os.makedirs(output_dir, exist_ok=True)
```

```python
# Función para guardar cada métrica en un archivo CSV
def save_metric_to_csv(tag):
    events = event_acc.Scalars(tag)
    metric_name = tag.replace("/", "_")  # Ajustar nombre del archivo según el tag
    csv_filename = os.path.join(output_dir, f"{metric_name}.csv")

    with open(csv_filename, mode='w', newline='') as csv_file:
        writer = csv.writer(csv_file)
        writer.writerow(["Wall time", "Step", "Value"])  # Encabezados

        # Escribir los datos en el CSV
        for event in events:
            writer.writerow([event.wall_time, event.step, event.value])

    print(f"Archivo CSV creado: {csv_filename}")
```

```python
# Recorrer y guardar cada métrica en su propio archivo CSV
for tag in tags:
    save_metric_to_csv(tag)
```
